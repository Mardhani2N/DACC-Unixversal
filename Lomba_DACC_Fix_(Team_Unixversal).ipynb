#### Setelah itu kami akan membuang data yang tidak diperlukan guna mempermudah mencari insight.

```python
# Membuang data yang tidak digunakan
cnn1 = cnn1.drop(columns=['Unnamed: 0'])
cnn1 = cnn1.drop(columns=['timezone'])
cnn1 = cnn1.drop(columns=['mentions'])
cnn1 = cnn1.drop(columns=['photos'])
cnn1 = cnn1.drop(columns=['hashtags'])
cnn1 = cnn1.drop(columns=['cashtags'])
cnn1 = cnn1.drop(columns=['link'])
cnn1 = cnn1.drop(columns=['reply_to'])
cnn1 = cnn1.drop(columns=['id'])
cnn1 = cnn1.drop(columns=['conversation_id'])
```

#Exploratory Data Analysis
### Exploratory Data Analysis (EDA) merupakan suatu proses menganalisis dataset untuk melihat karakteristik utama dari data. EDA adalah pemrosesan dasar yang dapat mempengaruhi keputusan-keputusan seorang data scientist dalam memproses data di tahap selanjutnya. Proses ini sering melibatkan dua hal, di antaranya descriptive statistics dan data visualization.

## Analyzing The Features

```python
cnn1['language']= cnn1['language'].str.replace('in','Indonesia') # Format awal, Format Baru
cnn1[['language']][:5]
```

### Merubah Tipe Data date yang awalnya object menjadi tipe data datetime.

```python
cnn1['date'] = cnn1['date'].astype('datetime64')

cnn1.info()
```

## Features Engineering

### Manipulasi Data Url untuk mengambil Topik untuk dijadikan Label.

```python
cnn2 = cnn1.copy()
```

```python
cnn2['urls'] = cnn2['urls'].str.replace('-', '')
cnn2['urls'].tail(6)
```

```python
label = cnn2['urls'].str.extract('(?:.*https://cnnindonesia.com/)([a-z]+)(?:.20)')
label.head(15)
```

```python
label.tail(15)
```

```python
cnn1[['label']] = label
cnn1.head()
```

```python
# Membuang Data Lama Yang Tidak Gunakan
cnn1 = cnn1.drop(columns=['urls'])
```

```python
# Melihat 5 data teratas setelah Membuang data lama
cnn1.head()
```

```python
cnn1.info()
```

#### Memindahkan columns Label yang berada diposisi terakhir menjadi disebelah columns tweet guna mempermudah mencari insight. 

```python
cnn1.insert(7, 'Label', cnn1['label'])
```

```python
cnn1 = cnn1.drop(columns=['label'])
```

```python
cnn1.head()
```

```python
cnn1.info()
```

## Insight Pie Chart Trending Topic CNN

```python
plt.figure(figsize=(15,8))

Jenis_Artikel = cnn1['Label'].value_counts()
labels = ['nasional', 'ekonomi', 'olahraga', 'internasional', 'teknologi', 'gayahidup','hiburan']

fig1, ax1 = plt.subplots()
ax1.pie(Jenis_Artikel, labels=labels, autopct='%1.1f%%', startangle=0)
ax1.axis('equal')

plt.title('Trending Topic CNN')
plt.pie(Jenis_Artikel)
plt.show()
```

## Observations :
### Bisa kita lihat bersama bahwa user CNN lebih banyak memilih topik Nasional dengan jumlah 30.770(34.4%), kemudiaan disusul dengan topik Ekonomi dengan jumlah 14.011(15.7%), dan topik yang kurang diminati oleh user CNN yaitu topik Hiburan dengan jumlah 6743 (7.5%).

## Insight Top Like Type of Topic

```python
# Mencari Insight
total_like = cnn1.groupby(['Label'])['likes_count'].sum().reset_index()
total_like
```

```python
# menganalisis perbandingan jumlah total like pada topik CNN
plt.figure(figsize=(15,8))
plt.bar(total_like['Label'],total_like['likes_count'],color=['green','red','blue','yellow','orange','purple','black'])
plt.title('Perbandingan Jumlah Total Like Pada Topik CNN')
plt.xlabel('Jenis Topik')
plt.ylabel('Total Like')
plt.show()
```
